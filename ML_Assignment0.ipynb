```python
from traitlets.config.application import deepcopy
import numpy as np
import matplotlib.pyplot as plt

# P1 Read files 
from google.colab import drive
drive.mount('/content/gdrive/', force_remount=True)

import cv2
img = cv2.imread('/content/gdrive/My Drive/Colab Notebooks/ML_2022_Fall/test.png')

# print(img)

# For each pixel in the image, i.e., every element in the array img,
# compute the Euclidean distance to the following points [255, 0, 0], [0, 255, 0] and [0, 0, 255]. 
# Once computed, set the value of the pixel to the point with the least distance. Save the image as output1.png
# Reference: https://blog.csdn.net/u013921430/article/details/106442429

# Euclidean distance Example
# dist = np.sqrt(np.sum(np.square(p1 - p2)))
# dist = np.linalg.norm(p1 - p2)

point1 = np.array([255, 0, 0])
point2 = np.array([0, 255, 0])
point3 = np.array([0, 0, 255])

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    p = img[i][j]
    dist1 = np.sqrt(np.sum(np.square(p - point1)))
    dist2 = np.sqrt(np.sum(np.square(p - point2)))
    dist3 = np.sqrt(np.sum(np.square(p - point3)))
    # print('dist 1:', dist1)
    # print('dist 2:', dist2)
    # print('dist 3:', dist3)
    
    m_dist = min(dist1, dist2, dist3)
    # print('min:', m_dist)

    if m_dist == dist1:
      img[i][j] = point1
    elif m_dist == dist2:
      img[i][j] = point2
    else:
      img[i][j] = point3

plt.figure()
plt.axis('off')
plt.imshow(img)
status = cv2.imwrite('/content/gdrive/My Drive/Colab Notebooks/ML_2022_Fall/output1.png', img)
print('Save image:', status)

######################################################################################3

# Set the pixels that fall within a square (whose sides are of length 50) 
# at the center of the image to be [0,0,0]. Save the image as output2.png.
# Reference: https://scikit-image.org/skimage-tutorials/lectures/00_images_are_arrays.html
# Reference: https://www.geeksforgeeks.org/python-opencv-cv2-imwrite-method/, https://www.w3schools.com/python/matplotlib_pyplot.asp

import cv2
img = cv2.imread('/content/gdrive/My Drive/Colab Notebooks/ML_2022_Fall/test.png')

new_img = np.array(img, copy=True)
height = new_img.shape[0]
width = new_img.shape[1]

c_h = int(height/2)
c_w = int(width/2)
new_img[c_h-50//2: c_h+50//2, c_w-50//2:c_w+50//2, :] = [0, 0, 0]
plt.figure()
plt.axis('off')
plt.imshow(new_img)
status = cv2.imwrite('/content/gdrive/My Drive/Colab Notebooks/ML_2022_Fall/output2.png', new_img)
print('Save image:', status)

```

Problem statement:
You and a friend go to buy tacos. You get three soft tacos and three burritos and your total bill is $11.25. Your friend's bill is $10.00 for four soft tacos and two burritos. How much do soft tacos cost? How much do burritos
cost?

```python
import numpy as np

# equation 
# 3*t + 3*b = 11.25
# 4*t + 2*b = 10
# t:1.25, b:2.5

# AX = B
# inv(A)A X= inv(A)dot B
# X = inv(A)dot B

A = np.array([[3, 3],[4,2]])
B = np.array([11.25, 10])
print(A)

print('Check Determinant ')
print(np.linalg.det(A))


## Matrix inversion
inv_A = np.linalg.inv(A) 
print(inv_A)

print('Method 1:')
X = inv_A.dot(B)
print('taco:', X[0], 'burrito:', X[1])

print('Method 2 using solve():')
X = np.linalg.solve(A,B)
print(X)

```
