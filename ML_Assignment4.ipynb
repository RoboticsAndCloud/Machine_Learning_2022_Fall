```python
# P(X1,X2 ,×××,Xn|C) = P(X1|C)P(X2|C)×××P(Xn|C)
min = 1E-50

APARTMENT = 'Apartment'
HOUSE = 'House'
CONDO = 'Condo'

def normpdf(x, mean, sd):
  var = float(sd)**2
  denom = (2*math.pi*var)**.5
  num = math.exp(-(float(x)-float(mean))**2/(2*var))
  return num/denom

def p_local_price(x, mu, sigma):
  prob = normpdf(x, mu, sigma)
  return prob

p_bathroom_dict = {APARTMENT:{1:5/7.0, 1.5:1/7.0, 2.5:1/7.0},
                   HOUSE:{1:6/7.0, 1.5:1/7.0, 2.5:0/7.0},
                   CONDO:{1:4/6.0, 1.5:1/6.0, 2.5:1/6.0},
                   }

p_garages_dict = {APARTMENT:{0:1/7.0, 1:3/7.0, 1.5:1/7.0, 2:2/7.0},
                   HOUSE:{1:2/7.0, 1:2/7.0, 1.5:1/7.0, 2:2/7.0},
                   CONDO:{0:0/6.0, 1:4/6.0, 1.5:0/6.0, 2:2/6.0},
                   }


p_bedrooms_dict = {APARTMENT:{2:1/7.0, 3:3/7.0, 4:2/7.0, 5:1/7.0},
                   HOUSE:{2:1/7.0, 3:5/7.0, 4:1/7.0, 5:0/7.0},
                   CONDO:{2:0/6.0, 3:5/6.0, 4:0/6.0, 5:1/6.0},
                   }

def p_bathroom(type, val):
  try: 
    p = p_bathroom_dict[type][val]
  except:
    p = min

  if p == 0:
    p = min

  return p

def p_land_area(x, mu, sigma):
  prob = normpdf(x, mu, sigma)
  return prob

def p_living_area(x, mu, sigma):
  prob = normpdf(x, mu, sigma)
  return prob

def p_garages(type, val):
  try: 
    p = p_garages_dict[type][val]
  except:
    p = min

  if p == 0:
    p = min

  return p

def p_rooms2(x, mu, sigma):
  prob = normpdf(x, mu, sigma)
  return min

def p_bedrooms(type, val):
  try: 
    p = p_bedrooms_dict[type][val]
  except:
    p = min

  if p == 0:
    p = min

  return p

def p_age_of_home(x, mu, sigma):
  prob = normpdf(x, mu, sigma)
  return min

# # P(X1,X2 ,×××,Xn|C) = P(X1|C)P(X2|C)×××P(Xn|C)
def bayes_prob(x_arr):
  P_apartment = p_local_price(x_arr[0], local_price_mu_apartment, local_price_sigma_apartment) \
   * p_bathroom(APARTMENT, x_arr[1]) \
   * p_land_area(x_arr[2], land_area_mu_apartment, land_area_sigma_apartment) \
   * p_living_area(x_arr[3], living_area_mu_apartment, living_area_sigma_apartment) \
   * p_garages(APARTMENT, x_arr[4]) * p_rooms(APARTMENT, x_arr[5]) \
   * p_bedrooms(APARTMENT, x_arr[6]) \
   * p_age_of_home(x_arr[7], age_home_mu_apartment, age_home_sigma_apartment)

  P_house = p_local_price(x_arr[0], local_price_mu_house, local_price_sigma_house) \
   * p_bathroom(HOUSE, x_arr[1]) \
   * p_land_area(x_arr[2], land_area_mu_house, land_area_sigma_house) \
   * p_living_area(x_arr[3], living_area_mu_house, living_area_sigma_house) \
   * p_garages(HOUSE, x_arr[4]) * p_rooms(HOUSE, x_arr[5]) \
   * p_bedrooms(HOUSE, x_arr[6]) \
   * p_age_of_home(x_arr[7], age_home_mu_house, age_home_sigma_house)

  P_condo = p_local_price(x_arr[0], local_price_mu_condo, local_price_sigma_condo) \
   * p_bathroom(CONDO, x_arr[1])\
   * p_land_area(x_arr[2], land_area_mu_condo, land_area_sigma_condo) \
   * p_living_area(x_arr[3], living_area_mu_condo, living_area_sigma_condo) \
   * p_garages(CONDO, x_arr[4]) * p_rooms(CONDO, x_arr[5]) \
   * p_bedrooms(CONDO, x_arr[6]) \
   * p_age_of_home(x_arr[7], age_home_mu_condo, age_home_sigma_condo)

  arr = [P_apartment, P_house, P_condo]
  arr_str = [APARTMENT, HOUSE, CONDO]
  print("P_apartment:", P_apartment)
  print("P_house:", P_house)
  print("P_condo:", P_condo)
  print("MAP:", " Classification ", arr_str[arr.index(max(arr))], ", prob:", max(arr))
  print("Accurate:", arr_str[arr.index(max(arr))] == x_arr[8], ", Expected:", x_arr[8])
  print("-------------------------------------------------")


def bayes_prob2(x_arr):
  P_apartment = p_local_price(x_arr[0], local_price_mu_apartment, local_price_sigma_apartment) \
   * p_bathroom(APARTMENT, x_arr[1]) \
   * p_land_area(x_arr[2], land_area_mu_apartment, land_area_sigma_apartment) \
   * p_living_area(x_arr[3], living_area_mu_apartment, living_area_sigma_apartment) \
   * p_garages(APARTMENT, x_arr[4]) * p_rooms2(x_arr[5], room_mu_apartment, room_sigma_apartment) \
   * p_bedrooms(APARTMENT, x_arr[6]) \
   * p_age_of_home(x_arr[7], age_home_mu_apartment, age_home_sigma_apartment)

  P_house = p_local_price(x_arr[0], local_price_mu_house, local_price_sigma_house) \
   * p_bathroom(HOUSE, x_arr[1]) \
   * p_land_area(x_arr[2], land_area_mu_house, land_area_sigma_house) \
   * p_living_area(x_arr[3], living_area_mu_house, living_area_sigma_house) \
   * p_garages(HOUSE, x_arr[4]) * p_rooms2(x_arr[5], room_mu_house, room_sigma_house) \
   * p_bedrooms(HOUSE, x_arr[6]) \
   * p_age_of_home(x_arr[7], age_home_mu_house, age_home_sigma_house)

  P_condo = p_local_price(x_arr[0], local_price_mu_condo, local_price_sigma_condo) \
   * p_bathroom(CONDO, x_arr[1])\
   * p_land_area(x_arr[2], land_area_mu_condo, land_area_sigma_condo) \
   * p_living_area(x_arr[3], living_area_mu_condo, living_area_sigma_condo) \
   * p_garages(CONDO, x_arr[4]) * p_rooms2(x_arr[5], room_mu_condo, room_sigma_condo) \
   * p_bedrooms(CONDO, x_arr[6]) \
   * p_age_of_home(x_arr[7], age_home_mu_condo, age_home_sigma_condo)

  arr = [P_apartment, P_house, P_condo]
  arr_str = [APARTMENT, HOUSE, CONDO]
  print("P_apartment:", P_apartment)
  print("P_house:", P_house)
  print("P_condo:", P_condo)
  print("MAP:", " Classification ", arr_str[arr.index(max(arr))], ", prob:", max(arr))
  print("Accurate:", arr_str[arr.index(max(arr))] == x_arr[8], ", Expected:", x_arr[8])
  print("-------------------------------------------------")

test1 = [6.0931,	1.5,	6.7265,	1.652,	1,	6,	3,	44,	'Apartment']
test2 = [8.3607,	1.5,	9.15,	1.777,	2,	8,	4,	48,	'House']
test3 = [8.14,	1,	8,	1.504,	2,	7,	3,	3,	'House']
test4 = [9.1416,	1.5,	7.3262,	1.831,	1.5,	8,	4,	31,	'Apartment']
test5 = [12,	1.5,	5,	1.2,	2,	6,	3,	30,	'Apartment']

bayes_prob2(test1)
bayes_prob2(test2)
bayes_prob2(test3)
bayes_prob2(test4)
bayes_prob2(test5)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix


from google.colab import drive
drive.mount('/content/gdrive/', force_remount=True)


import matplotlib.pyplot as plt
import numpy as np

# mode: decision tree
# data: dataset
def plot_feature_importances(model, data):
    n_features = data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), data.columns.values)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")

    plt.show()



xls = pd.ExcelFile('/content/gdrive/My Drive/Colab Notebooks/ML_2022_Fall/Assignment4/Asssignment4_Data.xlsx')
trainset = pd.read_excel(xls, 'Train')
testset = pd.read_excel(xls, 'Test')


# trainset.head()
# testset.head()
# print(trainset.columns.values)

X = trainset.drop('Construction type', axis=1)
X = X.drop('House ID', axis=1)
y = trainset['Construction type']

# print("X shape:",X.shape)
# print(X.columns.values)
# print(X.head())
# print('y:', y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print("Result of trainset")
print('Depth:', classifier.tree_.max_depth)
#print("Feature importance:",classifier.feature_importances_)

score = accuracy_score(y_test, y_pred)
score2 = classifier.score(X_test, y_test)
print("Accuracy: ", score, score2)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Get the importance of each feature
plot_feature_importances(classifier, X)


print("########################### Result of testset ################")
# Result of testset
X_test = testset.drop('Construction type', axis=1)
X_test = X_test.drop('House ID', axis=1)
y_test = testset['Construction type']
y_pred = classifier.predict(X_test)
print("Result of testset")
score = accuracy_score(y_test, y_pred)
score2 = classifier.score(X_test, y_test)
print("Accuracy: ", score, score2)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

```

```python
import numpy as np
import pandas as pd
from scipy.stats import mode
from sklearn.model_selection import train_test_split

# Euclidean Distance
def eucledian_distance(p1,p2):
    dist = np.sqrt(np.sum((p1-p2)**2))
    return dist
 
# KNN Algorithm
def KNN_predict(x_train, y_train , x_input, k):
    op_labels = []
     
    #Loop through the Datapoints to be classified
    for item in x_input: 
         
        #Array to store distances
        point_dist = []
         
        #Loop through each training Data
        for row in range(len(x_train)): 
            distances = eucledian_distance(np.array(x_train[row,:]) , item) 
            #Calculating the distance
            point_dist.append(distances) 
        point_dist = np.array(point_dist) 
         
        # Sorting the array while preserving the index
        # Keeping the first K datapoints
        dist = np.argsort(point_dist)[:k] 
         
        # Labels of the K datapoints
        labels = y_train[dist]
         
        # Majority voting, 
        #refer https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mode.html
        lab = mode(labels) 
        lab = lab.mode[0]
        op_labels.append(lab)
 
    return op_labels


from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from numpy.random import randint

# load data
from google.colab import drive
drive.mount('/content/gdrive/', force_remount=True)

# /content/drive/MyDrive/ExampleFile.txt
xls = pd.ExcelFile('/content/gdrive/My Drive/Colab Notebooks/ML_2022_Fall/Assignment4/Asssignment4_Data.xlsx')
trainset = pd.read_excel(xls, 'Train')
testset = pd.read_excel(xls, 'Test')

y = trainset['Construction type']
X = trainset.drop('Construction type', axis=1)
X = X.drop('House ID', axis=1)
X_train = X
y_train = y

X_test = testset.drop('Construction type', axis=1)
X_test = X_test.drop('House ID', axis=1)
y_test = testset['Construction type']

# Feature Scaling: ormalize the range of independent variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#Applying our function 
k = 3
y_pred = KNN_predict(X_train, y_train, X_test, k)
#Checking the accuracy
acc = accuracy_score(y_test, y_pred)

k_arr = []
acc_arr = []
for k in range(5):
  y_pred = KNN_predict(X_train, y_train, X_test, k+1)
  acc = accuracy_score(y_test, y_pred)

  k_arr.append(k+1)
  acc_arr.append(acc)

plt.figure(figsize=(10, 6))
plt.title('Train set')
plt.plot(k_arr, acc_arr, label='Accuracy', c='r')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
